```python
import pandas as pd
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter

# Load dataset
df = pd.read_csv('we_fn_use_c_marketing_customer_value_analysis.csv')

# Get unique states
states = df['State'].unique()

# Initialize geolocator
geolocator = Nominatim(user_agent="state_geocoder")
geocode = RateLimiter(geolocator.geocode, min_delay_seconds=1)

# Function to get latitude and longitude
def get_lat_lon(state_name):
    location = geocode(f"{state_name}, USA")
    if location:
        return pd.Series([location.latitude, location.longitude])
    else:
        return pd.Series([None, None])

# Create a dataframe with coordinates for each unique state
state_coords = pd.DataFrame(states, columns=['State'])
state_coords[['Latitude', 'Longitude']] = state_coords['State'].apply(get_lat_lon)

# Merge coordinates back to the original customer-level dataframe
df = df.merge(state_coords, on='State', how='left')

# Display updated dataframe with coordinates
display(df.head())

# save csv with coordinates
df.to_csv('customer_data_with_coordinates.csv', index=False)

# replace the split comma to ; in all the dataframe csv file
df.to_csv('customer_data_with_coordinates.csv', index=False, sep=';')
# save csv with coordinates
df.to_csv('customer_data_with_coordinates_replaced_semicolon.csv', index=False, sep=';')
```

📊 Regression Analysis: Customer Lifetime Value vs Income

🔹 Regression Equation  
Customer Lifetime Value = 0.00379 × Income + 7964.87  

---

🔹 Interpretation  

1. Relationship Between Variables  
The slope (0.00379) indicates a very weak positive relationship between Income and Customer Lifetime Value (CLV).  
For every €1 increase in Income, CLV increases by only €0.0038, which is negligible.  

---

2. Statistical Significance  
The p-value for Income is 0.221, which is greater than the standard threshold of 0.05.  
This indicates that the relationship is not statistically significant, meaning we cannot confidently conclude that Income affects CLV.  

---

3. Strength of the Model  
The very small slope and high p-value suggest that Income is a poor predictor of Customer Lifetime Value.  
There is little to no meaningful correlation between the two variables.  

---

4. Intercept Interpretation  
The intercept value (7964.87) suggests that even when Income is zero, the predicted CLV is approximately €7,965.  
This implies that other factors are driving customer value.  

---

🔹 Conclusion  
Overall, the regression analysis shows that Income has a negligible and statistically insignificant impact on Customer Lifetime Value.  
Other variables, such as customer behavior or policy characteristics, are likely more important in determining CLV.
